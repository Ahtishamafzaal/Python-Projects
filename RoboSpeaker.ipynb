```python
#importing the pacakges
from gtts import gTTS
from IPython.display import Audio

# defining a function
def speak(text):
    tts = gTTS(text=text, lang='en')
    tts.save('output.mp3')
    audio = Audio(filename='output.mp3')
    display(audio)

print("Hello, welcome to the Robo Speaker!")

#getting the input from the user
command = input("Enter the text you want to speak: ")

#calling the function
speak(command)
```
